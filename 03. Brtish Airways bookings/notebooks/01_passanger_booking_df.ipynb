### Calculating % sales channel

```python
booking_0_internet = df[(df['booking_complete'] == 0) & (df['sales_channel'] == 'Internet')].shape[0]
booking_0_mobile = df[(df['booking_complete'] == 0) & (df['sales_channel'] == 'Mobile')].shape[0]
```

```python
booking_1_internet = df[(df['booking_complete'] == 1) & (df['sales_channel'] == 'Internet')].shape[0]
booking_1_mobile = df[(df['booking_complete'] == 1) & (df['sales_channel'] == 'Mobile')].shape[0]
```

```python
total_bookings_internet = booking_0_internet + booking_1_internet
total_bookings_mobile = booking_0_mobile + booking_1_mobile
```

```python
percentage_booking_0_internet = (booking_0_internet / total_bookings_internet) * 100
percentage_booking_1_internet = (booking_1_internet / total_bookings_internet) * 100

percentage_booking_0_mobile = (booking_0_mobile / total_bookings_mobile) * 100
percentage_booking_1_mobile = (booking_1_mobile / total_bookings_mobile) * 100
```

```python
print(f"Percentage of Booking Complete = 0 for Internet: {percentage_booking_0_internet:.2f}%")
print(f"Percentage of Booking Complete = 1 for Internet: {percentage_booking_1_internet:.2f}%")
print(f"Percentage of Booking Complete = 0 for Mobile: {percentage_booking_0_mobile:.2f}%")
print(f"Percentage of Booking Complete = 1 for Mobile: {percentage_booking_1_mobile:.2f}%")
```
