```python
import statsmodels.tools.eval_measures as sm
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/dc_bikes_with_features_formatted_daytime.csv", parse_dates=['datetime'], index_col=0)
df.head()
```

```python
may_2011 = df[(df['year'] == 2011) & (df['month'] == 5)]['count']
may_2011
```

```python
may_2012 = df[(df['year'] == 2012) & (df['month'] == 5)]['count']
may_2012
```

```python
may_2011_table = pd.DataFrame(may_2011)
```

```python
may_2011_table
```

```python
may_2012_table = pd.DataFrame(may_2012)
may_2012_table
```

```python
sm.rmse(may_2011_table ['count'], may_2012_table ['count'])
```

# Conclusion:
The forecast on average differs from the real values by approx. 168 units.
