```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("../../03_kpi/data/dc_bikes_with_features_formatted_daytime.csv", parse_dates=['datetime'], index_col=0)
df.head()
```

```python
filtered_data = df[df['day'] == 5][df['month'] == 7][df['year'] == 2012]
filtered_data.head()
```

```python
sns.displot(data=filtered_data, x='count', kde=True, stat= 'probability')

plt.title("Distribution of rented bikes on 05.07.2012")
plt.xlabel("Count")
```

```python
filtered_2011 = df[df['year'] == 2011]
filtered_2011
```

```python
sns.displot(data=filtered_2011, x='temp', kde=True)

plt.title("Distribution of Temperature")
plt.xlabel("Temperature")
```
