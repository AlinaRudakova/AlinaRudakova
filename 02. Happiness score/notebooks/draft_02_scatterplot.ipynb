```python
countries_to_filter = ['Togo', 'South Africa', 'Turkey', 'Saudi Arabia', 'Romania', 'Denmark', 'Haiti', 'United States', 'Australia', 'New Zealand', 'Ecuador', 'Brazil']
year_to_filter = 2005
```

```python
filtered_df = happy[(happy['country'].isin(countries_to_filter)) & (happy['year'] == year_to_filter)]
```

```python
filtered_df
```

```python
sns.set_style('white')
plt.figure(figsize=(10, 6))

scatter = sns.scatterplot(data=filtered_df, x='year', y='happiness', s=70, hue= 'continent', alpha=0.7)
plt.xlabel('Year')
plt.ylabel('Happiness')
plt.title('Level of Happiness over the Years')
plt.grid(True)

scatter.set_xlim(left=2004, right=2020)
plt.ylim(25, 85)
plt.legend(loc = 'lower left', title='Continent', fontsize='small')
plt.xticks(ticks=range(2004, 2020), rotation=45)

for index, row in filtered_df.iterrows():
    plt.text(row['year'], row['happiness'], row['country'], ha='left', va='bottom', fontsize=8)
plt.show()
```

```python
years_range_df = happy[(happy['year'] >= 2004) & (happy['year'] <= 2020)]
years_range_df
```
