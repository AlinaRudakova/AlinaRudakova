```python
# Filter the DataFrame for the years 2015 to 2019
filtered_df = df[df['year'].between(2015, 2019)]

# Group by 'country' and apply nlargest to get the top 5 countries by happiness for each year
top_5_countries_2015_2019 = filtered_df.groupby('year').apply(lambda x: x.nlargest(5, 'happiness')).reset_index(drop=True)

# Create a line plot to show the happiness trend for each top 5 country over the years
plt.figure(figsize=(10, 6))
for country in top_5_countries_2015_2019['country'].unique():
    country_data = top_5_countries_2015_2019[top_5_countries_2015_2019['country'] == country]
    plt.plot(country_data['year'], country_data['happiness'], label=country)

plt.xlabel('Year')
plt.ylabel('Happiness')
plt.title('Happiness Trend in Top 5 Countries (2015-2019)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df = pd.read_csv('./all_data_merged.csv')
df
```

```python
df.describe()
```

```python
conditions = [
    (df['happiness'] < 44),
    (df['happiness'] >= 44) & (df['happiness'] < 62),
    (df['happiness'] >= 62)
]

labels = ['Unhappy Country', 'Neutral Country', 'Happy Country']

# Apply the conditions and labels using numpy.select() to create the 'happiness_category' column
df['happiness_category'] = np.select(conditions, labels)
```

```python
df
```

```python
population_count = df.groupby('happiness_category')['population'].sum().reset_index()
```

```python
population_count
```

```python
population_count.to_csv('pop_count.csv')
```
